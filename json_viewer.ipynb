```python
from IPython.display import display
import json
import pandas as pd

def json_to_table(path: str, sep=".") -> pd.DataFrame:
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)
    return pd.json_normalize(data, sep=sep)

df = json_to_table("test.json")
# print(df)

df_t = df.T.reset_index()
df_t.columns = ["column_name", "value"]

pd.set_option("display.max_rows", None)        # 行数の制限を解除
pd.set_option("display.max_columns", None)     # 列数の制限を解除
pd.set_option("display.width", 2000)           # 横幅を広げる
pd.set_option("display.max_colwidth", None)    # 列の文字数制限を解除
pd.set_option("display.expand_frame_repr", True)  # 省略せずに表示

display(df_t)


# display(df)
```
